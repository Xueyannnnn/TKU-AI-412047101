````python
import easyocr

from PIL import Image

from google.colab import files

from IPython.display import display, Markdown



def recognize_license_plate():

    # 提示使用者上傳照片

    print("請上傳您的車牌照片：")

    uploaded = files.upload()



    # 確認是否有文件被上傳

    if not uploaded:

        print("未上傳任何文件!")

        return



    # 對上傳的文件進行迭代

    for image_path_in_colab in uploaded.keys():

        # 開啟圖片並進行 OCR

        reader = easyocr.Reader(['en'])

        image = Image.open(image_path_in_colab)

        resce = reader.readtext(image, allowlist='ABCDEFGHIJKLMOPQRSPUVWXYZ1234567890-')



        # 提取識別結果

        extractedInformation = ""

        for result in resce:

            extractedInformation += result[1] + " "



        # 顯示結果

        display(Markdown(f"**識別結果從 {image_path_in_colab}:**"))

        display(Markdown(f"```\n{extractedInformation}\n```"))



        # 如果需要，可以選擇直接顯示圖片

        # display(image)



# 執行車牌辨識功能

recognize_license_plate()
````
